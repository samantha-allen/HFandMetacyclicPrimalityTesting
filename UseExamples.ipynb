## 1. A quick example: the trefoil

The trefoil's HF polynomial has no nontrivial positive-symmetric factorization, so this should resolve immediately at stage 1, before Jones/HOMFLY or anything downstream even runs.

```python
trefoil_pd = snappy.Link('3_1').PD_code()
result = HFandMetaPrimeTest(trefoil_pd, verbose=True)
result
```

## 2. DT code input

`HFandMetaPrimeTest` also accepts a DT code directly (with or without the `'DT: '` prefix) ŌĆö it's normalized to a PD code internally via SnapPy before anything else runs.

```python
result_dt = HFandMetaPrimeTest('DT: [(4,6,2)]', verbose=True)
result_dt
```

## 3. A case where the primeness-verification fails for prime covers up to 7: K13a3054 

```python
pd_K13a3054  = [[2, 0, 3, 25], [0, 14, 1, 13], [14, 2, 15, 1], [8, 3, 9, 4], [4, 20, 5, 19], [18, 6, 19, 5], 
                [6, 23, 7, 24], [24, 7, 25, 8], [20, 9, 21, 10], [10, 21, 11, 22], [16, 12, 17, 11], 
                [12, 16, 13, 15], [22, 17, 23, 18]]

result_K13a3054 = HFandMetaPrimeTest(pd_K13a3054, verbose=True)
result_K13a3054
```

## 4. Quick prime check with `is_prime_knot`

If you don't need the stage or the surviving factorizations, `is_prime_knot` just returns `True` (confirmed prime) or `None` (undetermined).

```python
print(is_prime_knot(trefoil_pd))
print(is_prime_knot(pd_K13a3054))
```

## 5. Going further: extending the prime list

If a knot comes back `inconclusive` with the default `primes=[2,3,5,7]`, one can repeat steps the homology and metacyclic tests with larger primes. The pattern below only pays the cost of the larger prime list when it's actually needed.

```python
def test_moreprimes(code, extra_primes=(11, 13, 17, 19, 23, 29)):
    result = HFandMetaPrimeTest(code, primes=[2, 3, 5, 7], verbose=True)
    if result['prime'] is None:
        print('undetermined at [2,3,5,7], retrying with', list(extra_primes), '...')
        result = HFandMetaPrimeTest(code, primes=[2, 3, 5, 7] + list(extra_primes), verbose=True)
    return result

test_moreprimes(pd_K13a3054)
```
